# Introdução

O objetivo deste trabalho é estimar os níveis de fecundidade no estado do Amazonas (AM) em 2010 a partir de dados dos Censo de 2000 e 2010, utilizando método de P/F de Brass (P/F-B) (Brass, 1974). Este método geralmente é utilizado em fontes de dados com qualidade razoável. 

Este método compara a fecundidade corrente acumulada com a parturição das mulheres e quando aplicado em informações censitárias corrigi erros de período de referência da informação (nos últimos 12 meses) relacionados a declarações incorretas na coleta de dados cometidas por valores culturais ou imprecisão na percepção do tempo, por exemplo. Quando aplicados a informações oriundas de registro civil, corrige erros de cobertura (sub-registro ou omissão).

Como pressupostos o método considera que, além de a qualidade dos dados ser razoável, eventuais erros de período de referência (a serem corrigidos pelo método) não possuem caráter seletivo com relação à idade das mães, ou seja, se existe o erro, acontece de maneira similar para todas as faixas etárias. Outro pressuposto é que não existe incidência de erros de "memória ausente" (declaração de número de filhos menor que real), o que poderia subestimar a fecundidade observado nos dados. Por fim, são considerados também população fechada, mortalidade não seletiva e Taxas Específicas de Fecundidade (TEFs) constantes entre as mulheres em idade reprodutiva.

Os insumos de dados necessários para aplicação do método P/F-B são:

* **Pop**: população feminina;
* **P**: parturição, i.e., número de filhos tidos por mulher na data de referência;
* **F**: fecundidade corrente, i.e., número de filhos nascidos-vivos nos últimos 12 meses;

A fecundidade (f) é corrigida pela fator de correção P/F, onde P representa o nível de fecundidade e é dado pela parturição (P) observada (de uma determinada coorte); e F é a estrutura da fecundidade é obtida pela distribuição das TEFs. Com isso, corrige-se a fecundidade de forma similar aos que é feito nos métodos de padronização indireta, com a seguinte fórmula:

    f(ajustada) = f(observada) * P/F

## Passo 1: Estimando parturições médias e fecundidade corrente

Para realizar ajuste utilizando o P/F-B, a primeira etapa consiste em calcular estimativas de parturição média (P) e fecundidade corrente (fi) para os dados observados. Nas Tabelas 5 e 6 estão apresentados os resultados dos cálculos de *P* e de *fi*, no estado do AM para os anos de 2000 e 2010 segundo dados do Censo, os quais estão representados na Figura 2. Na Figura 2a podemos observar que a parturição média aumenta de maneira gradativa conforme avança-se nas faixas etárias, alcançando valor próximo de 5 ao final da vida reprodutiva em 2000, e menor que 4 em 2010. Também estão representados nesta figura valores observados de *P* e de *fi* utilizando dados das PNADs 2001 e 2011, entretendo o problema de cobertura fica mais uma vez evidenciado, tanto em 2001 quanto em 2011, indicando valores inferiores a 1 nos dois casos.

Na Figura 2b, nota-se de maneira clara uma queda na fecundidade em todas as faixas etárias de 2000 para 2010; e na Figura 2a nota-se a queda na parturição média em todas as faixas etárias de 2000 para 2010.


```r
passo1 <- function(df) {
    # P: parturição (estimanda)
    df$P = round(df$nasc_v / df$mulheres, 5)

    # fi: Estimando a taxa de fecundidade corrente ###
    df$fi = round(df$nasc_12 / df$mulheres, 5)
    return(df)
}

# Censo ----
df2000 = passo1(df2000)
TFT_2000 = round(sum(df2000$fi) * 5,3)

df2010 = passo1(df2010)
TFT_2010 = round(sum(df2010$fi) * 5, 3)

# PNAD ----
dfPNAD2001 = passo1(dfPNAD2001)
TFT_PNAD_2001 = round(sum(dfPNAD2001$fi) * 5, 3)

dfPNAD2011 = passo1(dfPNAD2011)
TFT_PNAD_2011 = round(sum(dfPNAD2011$fi) * 5, 3)
```

## Passo 2: Estimando parâmetros auxiliares 

A segunda etapa da execução do método P/F-B consiste em calcular Idade Média a Fecundidade e a razão *P1/P2*, sendo *P1* o valor observado da parturição do grupo etário de 15 a 19 anos, e P2 o valor observado da parturição do grupo etário de 20 a 24 anos. Estes "parâmetros auxiliares" são utilizados como referência no processo de definição dos valores de ajuste (k) a serem utilizados posteriormente para correção. Com eles, os valores de k são calculados interpolando-os com os valores imediatamente inferior e superior da tabela padrão de fecundidade (Tabela 8) proposta por Brass e Coale (1968 apud Carvalho, 2018). A Tabela 7 apresenta as Taxa de Fecundidade Total (TFT), as Idade Média à Fecundidade (IMF) e as razão P1/P2, para o estado do AM nos anos de 2000 e 2010 segundo Censos, e 2001 e 2011 segundo PNAD.


```r
# Idade média a fecundidade - "ponto médio"
idades = seq(15,45,5)
df2000$pont_m = idades + 2.5

#cat("Idade Média à fecundidade:", idad_m)
idad_m_2000 = sum(df2000$pont_m * df2000$fi)/sum(df2000$fi)

#cat("\nRazão entre os dois primeiros Ps: ", razao_p)
razao_p_2000 = df2000$P[1]/df2000$P[2]

# 2010 ----------------------------------------------
df2010$pont_m = idades + 2.5

#cat("Idade Média à fecundidade:", idad_m)
idad_m_2010 = sum(df2010$pont_m * df2010$fi)/sum(df2010$fi)

#cat("\nRazão entre os dois primeiros Ps: ", razao_p)
razao_p_2010 = df2010$P[1]/df2010$P[2]


# PNAD ----------------------------------------------
dfPNAD2001$pont_m = idades + 2.5

#cat("Idade Média à fecundidade:", idad_m)
idad_m_2001_PNAD = sum(dfPNAD2001$pont_m * dfPNAD2001$fi)/sum(dfPNAD2001$fi)

dfPNAD2011$pont_m = idades + 2.5

#cat("Idade Média à fecundidade:", idad_m)
idad_m_2011_PNAD = sum(dfPNAD2011$pont_m * dfPNAD2011$fi)/sum(dfPNAD2011$fi)
```

**Tabela 7.** *Taxa de Fecundidade Total (TFT), Idade Média à Fecundidade (IMF) e razão P1/P2, para o estado do AM nos anos de 2000 e 2010 segundo dados dos censos anos 2000 e 2010.*

```r
dfParametros <- data.frame(col1 = c("TFT", "IMF", "P1/P2"), 
                           col2 = c(round(TFT_2000,4), round(idad_m_2000,2), round(razao_p_2000,4)), 
                           col3 = c(round(TFT_2010,4), round(idad_m_2010,2), round(razao_p_2010,4)),
                           col4 = c(round(TFT_PNAD_2001,4), round(idad_m_2001_PNAD,2), 0),
                           col5 = c(round(TFT_PNAD_2011,4), round(idad_m_2011_PNAD,2), 0))


colnames(dfParametros) <- c("Parâmetro", "Censo 2000", "Censo 2010", "PNAD 2001", "PNAD 2011")
dfParametros 
```

## Passo 4: Cálculo, por interpolação, dos multiplicadores "k" para ajuste 

Identificados os intervalos realiza-se então os cálculos dos valores de k a serem utilizados no ajuste. A Figura 4 apresenta de forma gráfica os valores já interpolados, a serem utilizados no ajuste, para os anos de 2000 e 2010, assim como os valores padrões de cada intervalo da tabela padrão.


```r
# Alfa para os 3 primeiros grupos etários
# Selecionando os valores da tabela de multiplicadores para os intervalos encontrados (int5 - int6).
mult_3 = mult[1:3, 6:7]

k_p1_p2_inf = mult[8:8, 6:6]
k_p1_p2_sup = mult[8:8, 7:7]

# interpolação
alfa_2000 = (razao_p_2000 - k_p1_p2_inf) / (  k_p1_p2_sup - k_p1_p2_inf) 
alfa_2010 = (razao_p_2010 - k_p1_p2_inf) / (  k_p1_p2_sup - k_p1_p2_inf) 

# k = alfa * (intervalo superior) - alfa * (intervalo inferior) + (intervalo inferior)
mult_3$ks_2000 = (alfa_2000 * (mult_3$Int6)) - (alfa_2000 * (mult_3$Int5)) + (mult_3$Int5)
mult_3$ks_2010 = (alfa_2010 * (mult_3$Int6)) - (alfa_2010 * (mult_3$Int5)) + (mult_3$Int5)

# Para os 4 últimos grupos etários ------------------------------------------
mult_4_2000 = mult[4:7, 7:8] # (int6 : int7)
mult_4_2010 = mult[4:7, 6:7] # (int5 : int6)

# interpolação ano 2000 --
k_IMF_inf = mult[10:10, 7:7]
k_IMF_sup = mult[10:10, 8:8]

alfa_n_2000 = (idad_m_2000 - k_IMF_inf) / (k_IMF_sup - k_IMF_inf)

# interpolação ano 2010 --
k_IMF_inf = mult[10:10, 6:6]
k_IMF_sup = mult[10:10, 7:7]

alfa_n_2010 = (idad_m_2010 - k_IMF_inf) / (k_IMF_sup - k_IMF_inf)

# K para os últimos grupos
#alfa(intervalo inferior)-alfa(intervalo superior)+(intervalo superior)
mult_4_2000$ks_2000 = alfa_n_2000 * (mult_4_2000$Int6) - alfa_n_2000 * (mult_4_2000$Int7) + (mult_4_2000$Int7)
mult_4_2010$ks_2010 = alfa_n_2010 * (mult_4_2010$Int5) - alfa_n_2010 * (mult_4_2010$Int6) + (mult_4_2010$Int6)

# Combinando os multiplicadores num único vetor
df2000$ki = c(mult_3$ks_2000, mult_4_2000$ks_2000)
df2010$ki = c(mult_3$ks_2010, mult_4_2010$ks_2010)
```

**Figura 4.** *Gráficos dos valores de ajuste (k) calculados por interpolação com valores da tabela padrão (os quais serão utilizados para corrigir dados de fecundidade do AM 2000 e 2010).*

## Passos 5 e 6: Estimativa de P/F e realização do ajuste dos níveis de fecundidade

As próximas duas etapas do processo consistem na correção da parturição média utilizando os valores de k estimados, seguido da estimativa da razão P/F e realização do ajuste dos níveis de fecundidade. Os resultados dos cálculos estão apresentados nas Tabelas 10 e 11, para os anos de 2000 e 2010 respectivamente.

A Figura 5 apresenta os resultados da fecundidade já ajustada para os dois anos analisados, 2000 na Figura 5a, e 2010 na Figura 5b. Estão apresentados também os valores da fecundidade observada (inicialmente). Como podemos notar, em ambos os anos analisados, a fecundidade observada apresenta níveis menores em todos os grupos etários, mostrando uma maior discrepância para os grupos etários iniciais, e valores mais próximos (entre observado e ajustado) nas duas últimas faixas etárias (de 40 a 49 anos).


```r
passo5 <- function(df) {
    df$delta = 0
    for (i in 1:6) {
      df$delta[i+1] = round((df$fi[i]) * 5 + (df$delta[i]),5)
    }

    # Fi = Parturição média corrigida
    df$Fi = round(df$delta + df$ki * df$fi,5)

    # P/F
    df$P_F = round(df$P/df$Fi,5)

    # Passo 6: ajuste utilizando grupo de ajuste, 20-24 anos.

    df$fi_aj = round(df$P_F[2] * df$fi, 5)
    return(df)
}

df2000 = passo5(df2000)
df2010 = passo5(df2010)
```

## Passos 7 e 8: Ajuste de 6 meses e cálculo das TFTs ajustadas

Por as últimas etapas do processo são a realização do ajuste dos 6 meses entre o tempo de nascimentos e cálculo das TFTs ajustadas pelo P/F-B já com este ajuste de 6 meses. Nos Quadros 1 e 2 estão apresentadas as TFTs observadas e corrigidas com o ajuste de 6 meses para os anos de 2000 e 2010 respectivamente. As TEFs ajustadas estão apresentadas nas Tabelas 12 e 13 para os anos de 2000 e 2010 respectivamente.
Para o ano de 2000, a TFT foi ajustada de 3,02 para 3,45 filhos por mulher. Já em 2010, a correção foi menor, porém também foi observado uma maior TFT (de 2,22 para 2,66 filhos por mulher).

As Figuras 6a e 7a apresentam gráficos da TEFs observadas, corrigidas e também observadas da PNAD, para os anos de 2000 e 2010 respectivamente. Como podemos observar as TEFs corrigidas apresentam níveis maiores que as observadas em todos os grupos etários. No gráfico 7a fica evidenciado que segundo a PNAD 2011 a TEF na faixa etária de 30 a 34 anos apresenta nível maior que a do Censo, corrigida inclusive. Na faixa etária seguinte, de 35 a 39 anos, apresenta nível entre a observada e a corrigida pelo Censo, enquanto era esperado que os níveis fossem inferiores à observada, seguindo o padrão do Censo, que possui uma cobertura bem mais ampla.

Logo após a divulgação dos dados censitários de 2010, houve muita discussão sobre a qualidade da informação sobre fecundidade, em função da constatação de que os fatores de ajustes (razões P/F) tinham aumentado em comparação, principalmente, ao Censo Demográfico de 2000 (Cavenaghi, 2006). Para o estado do AM isto é observado, e como pode ser visto na Figura 7b, os fatores de ajustes são menores em 2000 que em 2010 para todas as faixas etárias, exceto para a última. Além disso, todos os valores são maiores do que um, o que significa que a fecundidade corrente está subestimada ou está declinando, e no caso sabe-se que está em declínio, principalmente nas faixas mais avançadas do período reprodutivo. Por fim, nas duas últimas faixas etárias o ajuste realizado foi em torno de 50%.


```r
# criando os fatores de ajuste para dados observados e corrigidos
fobs_aj_2000 = c(
  (df2000$fi[1] + df2000$fi[2])/20,
  (df2000$fi[3] - df2000$fi[1])/20,
  (df2000$fi[4] - df2000$fi[2])/20,
  (df2000$fi[5] - df2000$fi[3])/20,
  (df2000$fi[6] - df2000$fi[4])/20,
  (df2000$fi[7] - df2000$fi[5])/20,
  (-df2000$fi[6] - df2000$fi[7])/20
  )

fest_aj_2000 = c(
  (df2000$fi_aj[1] + df2000$fi_aj[2])/20,
  (df2000$fi_aj[3] - df2000$fi_aj[1])/20,
  (df2000$fi_aj[4] - df2000$fi_aj[2])/20,
  (df2000$fi_aj[5] - df2000$fi_aj[3])/20,
  (df2000$fi_aj[6] - df2000$fi_aj[4])/20,
  (df2000$fi_aj[7] - df2000$fi_aj[5])/20,
  (-df2000$fi_aj[6] - df2000$fi_aj[7])/20
  )

# 2010
fobs_aj_2010 = c(
  (df2010$fi[1] + df2010$fi[2])/20,
  (df2010$fi[3] - df2010$fi[1])/20,
  (df2010$fi[4] - df2010$fi[2])/20,
  (df2010$fi[5] - df2010$fi[3])/20,
  (df2010$fi[6] - df2010$fi[4])/20,
  (df2010$fi[7] - df2010$fi[5])/20,
  (-df2010$fi[6] - df2010$fi[7])/20
  )

fest_aj_2010 = c(
  (df2010$fi_aj[1] + df2010$fi_aj[2])/20,
  (df2010$fi_aj[3] - df2010$fi_aj[1])/20,
  (df2010$fi_aj[4] - df2010$fi_aj[2])/20,
  (df2010$fi_aj[5] - df2010$fi_aj[3])/20,
  (df2010$fi_aj[6] - df2010$fi_aj[4])/20,
  (df2010$fi_aj[7] - df2010$fi_aj[5])/20,
  (-df2010$fi_aj[6] - df2010$fi_aj[7])/20
  )
```

**Quadro 1.** *TFTs observada e corrigida com o ajuste de 6 meses em 2000 segundo Censo.*

```r
resultado_2000 = data.frame(idade = cbind(idade=seq(15,45,5)))
resultado_2000$fi = round(df2000$fi + fobs_aj_2000, 4)
resultado_2000$fi_aj = round(df2000$fi_aj + fest_aj_2000, 4)

## Estimando TFTs observada e corrigida com o ajuste de 6 meses ###
TFT_n_2000 = sum(resultado_2000$fi) * 5
TFT_c_2000 = sum(resultado_2000$fi_aj) * 5

cat("\nTFT observada: ", TFT_n_2000)
cat("\nTFT corrigida com ajuste de 6 meses: ", TFT_c_2000, "\n\n")
```
