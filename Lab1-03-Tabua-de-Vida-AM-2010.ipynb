```python
dfTabuaMulheres = pd.read_csv("dados/Tabua-Mulheres-2010-AM.csv")
dfTabuaMulheres
```

```python
dfTabuaWL21M = pd.read_csv("dados/tabuas/CoaleDemeny-west-Level-21-M.csv")
dfTabuaWL21M.drop(columns=['Unnamed: 0'], inplace=True)

dfTabuaWL22M = pd.read_csv("dados/tabuas/CoaleDemeny-west-Level-22-M.csv")
dfTabuaWL22M.drop(columns=['Unnamed: 0'], inplace=True)

dfTabuaWL23M = pd.read_csv("dados/tabuas/CoaleDemeny-west-Level-23-M.csv")
dfTabuaWL23M.drop(columns=['Unnamed: 0'], inplace=True)

dfTabuaWL24M = pd.read_csv("dados/tabuas/CoaleDemeny-west-Level-24-M.csv")
dfTabuaWL24M.drop(columns=['Unnamed: 0'], inplace=True)

dfTabuaWL25M = pd.read_csv("dados/tabuas/CoaleDemeny-west-Level-25-M.csv")
dfTabuaWL25M.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
print("Expectativa de vida ao nascer:")
print("Observado: ", dfTabuaHomens["ex"][0])
print("West-L-21: ", round(dfTabuaWL21M["ex"][0],2))
print("West-L-22: ", round(dfTabuaWL22M["ex"][0],2))
print("West-L-23: ", round(dfTabuaWL23M["ex"][0],2))
print("West-L-24: ", round(dfTabuaWL24M["ex"][0],2))
print("West-L-25: ", round(dfTabuaWL25M["ex"][0],2))
```

```python
lx = dfTabuaWL24M.copy()
lx.drop(columns={"lx","nqx","dx","mx","Lx","Px","Tx","ex"}, inplace=True)

lx["l(x)"] = dfTabuaHomens["lx"]

lx["21"] = dfTabuaWL21M["lx"]
lx["22"] = dfTabuaWL22M["lx"]
lx["23"] = dfTabuaWL23M["lx"]
lx["24"] = dfTabuaWL24M["lx"]
lx["25"] = dfTabuaWL25M["lx"]

lx.rename(columns={"age":"Idade"}, inplace=True)
lx
```

### Conjunto de interpolacoes para encontrar meu nível na tabua modelo 
X = (lxestimado - lxinferior)/(lxsuperior-lxinferior) + intervalo inferior (valor do nível)


```python
# Valores grupos etários 5-10 anos, 20-25 anos, e 40-45 anos
print(lx.iloc[[2]])
print(lx.iloc[[5]])
print(lx.iloc[[9]])
```

```python
# Usando grupos etários 5-10 anos, 20-25 anos, e 40-45 anos
def interpolar(lx, row, level):
    lxestimado = lx.iloc[[row]]["l(x)"]
    lxinferior = lx.iloc[[row]][str(level)]
    lxsuperior = lx.iloc[[row]][str(level + 1)]
    
    return float((lxestimado - lxinferior)/(lxsuperior-lxinferior) + level)

ponto5 = interpolar(lx, 2, 23)
ponto20 = interpolar(lx, 5, 22)
ponto40 = interpolar(lx, 9, 21)

print("Nível escolhido para ser utilizado para ajuste: " , round(np.mean([ponto5, ponto20, ponto40]),2))

lx.drop(columns={"21","24","25"}, inplace=True)
```

```python
lx["ls(x)"] = (lx["23"]-lx["22"]) * 0.67 + lx["22"]
lx
lx.to_csv("dados/lx-2010-AM-Homens.csv", sep=";", index=False)
```

```python
# Gráfico TEM Grupos etários > 1 ano e < 80
fig, ax = plt.subplots(figsize=(14,7))
ax.set(xlabel='Grupos Etários', ylabel='lx',
       title='Funções de Sobrevivência do AM-2010-Homens e das tábuas modelo.');

x = lx["Idade"]
ax.set_xticks(x)

plt.plot(x, lx["l(x)"], linestyle='solid', label="l(x) tábua original AM-2010")
plt.plot(x, lx["22"], linestyle='dashed', label="l(x) da tábua modelo nível 22")
plt.plot(x, lx["23"], linestyle='dashdot', label="l(x) da tábua modelo nível 23")
plt.plot(x, lx["ls(x)"], linestyle='dotted', label="ls(x) interpolado no nível %s" % round(np.mean([ponto5, ponto20, ponto40]),2))

plt.xticks(rotation=45)
plt.grid(True)

plt.legend();
```
