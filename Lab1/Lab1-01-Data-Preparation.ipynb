## Ajustes e Correções
- Distribuição proporcional dos Ignorados (pro-rata)
- Agrupamento dos óbitos das faixas com mais de 80 anos na faixa 80+. Esta abordage foi utilizada considerando os dados SIM estão agregados por idade até a faixa de 80+, assim será feito este mesmo ajuste para Censo e para Registro Civil

```python
# Função de distribuição proporcional dos ignorados
def ign_pro_rata_ditro(dfAux):
    for col in ["Homens", "Mulheres"]:
        # formula distribuição pro_rata: (obitos da faixa etária) + (Ignorados/total de óbitos) * obitos da faixa etária)
        dfAux[col] = dfAux[col] + (dfAux[col].iloc[22]/dfAux[col].sum()) * dfAux[col]
        dfAux[col] = round(dfAux[col],1)
        dfAux[col].iloc[22] = 0

# SIM possui ignorados na faixa "Menos de 1 ano"
homens = dfObitosSIM["Homens"].iloc[0]
mulheres = dfObitosSIM["Mulheres"].iloc[0]
ign = dfObitosSIM["Ignorados"].iloc[0]
tot = mulheres + homens
dfObitosSIM["Homens"].iloc[0] = homens + (ign/tot) * homens
dfObitosSIM["Mulheres"].iloc[0] = mulheres + (ign/tot) * mulheres

ign_pro_rata_ditro(dfObitosRegistroCivil)
ign_pro_rata_ditro(dfObitosCenso)
ign_pro_rata_ditro(dfObitosSIM)

```

```python
# Função para agrupamento de óbitos na faixa 80+
def agruparIdades(dfAux):
    for col in ["Homens", "Mulheres"]:
        dfAux[col].iloc[17] =  dfAux[col].iloc[17] + dfAux[col].iloc[18] + dfAux[col].iloc[19] + dfAux[col].iloc[20] + dfAux[col].iloc[21]
        dfAux['Idade'].iloc[17] = "80+"

#
#agruparIdades(dfObitosRegistroCivil)
#agruparIdades(dfObitosCenso)
#agruparIdades(dfObitosSIM)
#agruparIdades(dfPop)

# Remoção de linhas e colunas
#dfObitosRegistroCivil = dfObitosRegistroCivil.drop([18,19,20,21,22])
#dfObitosCenso = dfObitosCenso.drop([18,19,20,21,22])
#dfObitosSIM = dfObitosSIM.drop([18,19,20,21,22])
#dfPop = dfPop.drop([18,19,20,21,22])
```
